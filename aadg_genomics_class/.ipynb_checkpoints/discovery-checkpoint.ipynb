```python
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
with open('../discovery_full.json', 'r') as data_file:
    data = json.load(data_file)
```

```python
#train|test.nucl_probs_pos_per_class
```

```python
test_classes = dict()
got_header = False
with open("../testing_ground_truth.tsv") as gt_file:
    for line in gt_file:
        if got_header:
            split_line = line.split("\t")
            if len(split_line) < 2:
                continue
            [fasta_file_gz, assigned_class, *_] = split_line
            test_classes[fasta_file_gz.strip()] = assigned_class.strip()
        got_header = True

def prob_dist(a, b):
    c = []
    for nucl in set([*list(a.keys()), *list(b.keys())]):
        if nucl != "N":
            a0 = np.array(a[nucl])
            b0 = np.array(b[nucl])
            c.append(np.linalg.norm(a0-b0))
    return np.linalg.norm(np.array(c))

test_ds_dists = []
for (dataset_name, ds) in data["test"]["datasets"].items():
    cls = test_classes[dataset_name.replace("/home/staff/iinf/ajank/adg/", "")]
    probs = ds["nucl_probs_pos"]
    dist_to_ok = prob_dist(probs, data["train"]["nucl_probs_pos_per_class"][cls])
    dist_to_bad = min([prob_dist(probs, probs2) for (cls2, probs2) in data["train"]["nucl_probs_pos_per_class"].items() if cls2 != cls])
    test_ds_dists.append((dataset_name, dist_to_ok, dist_to_bad))

test_ds_dists
```
